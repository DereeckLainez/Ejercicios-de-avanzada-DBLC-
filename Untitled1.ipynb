```python
# desarrollar una aplicación para convertir entre monedas (Lempiras, dólares, euros)

def convertir_moneda(cantidad, moneda, precios):
    print("ingrese la moneda que desea convertir lempiras, dolares, euros:")
    moneda = input().lower()
    if moneda not in precios:
        print("Moneda no válida.")
        return None
    if moneda == "lempiras":
        dolares = cantidad / precios["dolares"]
        euros = cantidad / precios["euros"]
        print(f"{cantidad} Lempiras son {dolares:.2f} Dólares y {euros:.2f} Euros.")
    elif moneda == "dolares":
        lempiras = cantidad * precios["lempiras"]
        euros = cantidad / precios["euros"]
        print(f"{cantidad} Dólares son {lempiras:.2f} Lempiras y {euros:.2f} Euros.")
    elif moneda == "euros":
        lempiras = cantidad * precios["lempiras"]
        dolares = cantidad * precios["dolares"]
        print(f"{cantidad} Euros son {lempiras:.2f} Lempiras y {dolares:.2f} Dólares.")
```
