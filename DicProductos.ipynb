```python
#Compra de un producto

#diccionario de valores

producto={"Telefono":3000,"Descuento":0.15,"Impuesto":0.30}
Cliente=input("Ingrese su nombre: ")
print("Bienvenido", Cliente)
print("El precio del producto es:", producto["Telefono"])
precio=producto["Telefono"]
descuento=producto["Descuento"]
impuesto=producto["Impuesto"]
precio_con_descuento=precio-(precio*descuento)
precio_con_impuesto=precio_con_descuento+(precio_con_descuento*impuesto)
print("El precio con descuento es:", precio_con_descuento)
print("El precio con impuesto es:", precio_con_impuesto)
respuesta=input("¿Desea comprar el producto? (si/no): ")
if respuesta.lower() == "si":
    print("Gracias por su compra, ", Cliente)
else:
    print("Gracias por visitarnos, ", Cliente)
# Fin del programa
```
